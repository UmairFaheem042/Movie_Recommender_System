```python
# Desired format of genres => ['Action', 'Adventure', 'Fantasy', 'Science Fiction']
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# ['Action', 'Adventure', 'Fantasy', 'Science Fiction']
import ast
def convert3(obj):
    counter = 0
    L = []
    for i in ast.literal_eval(obj):
        if counter != 3: # get only first three cast name
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['cast'][0]
```

```python
movies.head(2)
```

```python
movies['crew'][0] # requires only director
```

```python
def fetchDirector(obj):
    L = []
    for i in ast.literal_eval(obj):
        # crew as key as job where value is Editor/Director/Sound_Designer/Casting
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetchDirector)
```

```python
movies['crew'][0]
```

```python
movies.head(2)
```

```python
# convert overview to a list too
movies['overview'] = movies['overview'].apply(lambda x: x.split()) 
```

```python
movies['overview'][0]
```

```python
movies.head(1)
```

```python
# remove spaces in words(cast/crew/keywords...) because if cast = Sam Worthington is made tag it will be 'Sam' & 'Worthington' but what if 'Sam' is also someone esle name then it would be a problem. Therefore remove space 'SamWorthington' so that tag is made this.
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head(2)
```

```python
# new column added which is a combination of overview + genres + keywords + cast + crew
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0] # avatar
```

```python
# stemming => have learned about this in 5th sem (AI)
# ['loved', 'love', 'loving'] => ['love', 'love', 'love']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# ps.stem('loving') # => love
# ps.stem('love') # => love

ps.stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem) # look at marin it is being stemmed from marine
```

```python
new_df['tags'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# one movie's distance is calculated with all movies i.e. 4806 with 4806 movies
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0] # problem with this is that when we will sort the values will be sorted not the index so we won't be able to know which movie was it.
list(enumerate(similarity[0])) # this will create a list of tuples with index and similarity values
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1]) # sorted in reverse order on the basis of 2nd value
```

```python
def recommend(movie):
    # we will get movie (user selected) and then similar vectors, after we get vectors from it, now we will sort it.
    # after sorting we will fetch first n(we will take 5) movies.
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:11] #=> 1:6 is top 5

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
import string
entered_movie = string.capwords('INside out')
recommend(entered_movie)
#recommend('Inside Out')
```
